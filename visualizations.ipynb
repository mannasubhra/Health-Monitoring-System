#Required Libraries & Data upload

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")


df = pd.read_csv("patients_data.csv")
sns.set_style("whitegrid")
```

#Visualizations

```python
numeric_df = df.select_dtypes(include=["number"])

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Patient Health Metrics")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# BMI Distribution
sns.histplot(df["BMI"], bins=30, kde=True, ax=axes[0, 0], color="blue")
axes[0, 0].set_title("BMI Distribution")

# Blood Pressure Distribution
sns.histplot(df["BP_Systolic"], bins=30, kde=True, ax=axes[0, 1], color="red")
axes[0, 1].set_title("Systolic Blood Pressure Distribution")

# Sugar Level Distribution
sns.histplot(df["Sugar_Level"], bins=30, kde=True, ax=axes[1, 0], color="green")
axes[1, 0].set_title("Sugar Level Distribution")

# Cholesterol Distribution
sns.histplot(df["Cholesterol"], bins=30, kde=True, ax=axes[1, 1], color="purple")
axes[1, 1].set_title("Cholesterol Level Distribution")

plt.tight_layout()
plt.show()
```

```python
#Boxplots to Detect Outliers
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(data=df[["BMI", "BP_Systolic", "Sugar_Level", "Cholesterol"]])
plt.title("Boxplots of Key Health Metrics")
plt.xticks(rotation=45)
plt.show()
```

```python
# BMI Category Count
df["BMI_Category"] = df["BMI"].apply(lambda x: "Underweight" if x < 18.5 else "Normal" if x < 25 else "Overweight")

plt.figure(figsize=(8, 5))
sns.countplot(x="BMI_Category", data=df, palette="Set2")
plt.title("Count of Patients by BMI Category")
plt.xlabel("BMI Category")
plt.ylabel("Count")
plt.show()
```

```python
# Blood Pressure Classification Count
df["BP_Category"] = df["BP_Systolic"].apply(lambda x: "Normal" if x < 120 else "High BP (Hypertension)")

plt.figure(figsize=(8, 5))
sns.countplot(x="BP_Category", data=df, palette="coolwarm")
plt.title("Count of Patients by Blood Pressure Category")
plt.xlabel("BP Category")
plt.ylabel("Count")
plt.show()
```

```python
# Diabetes Risk Based on Sugar Levels
df["Sugar_Risk"] = df["Sugar_Level"].apply(lambda x: "Normal Sugar Level" if x < 140 else "Diabetes Risk")

plt.figure(figsize=(8, 5))
sns.countplot(x="Sugar_Risk", data=df, palette="muted")
plt.title("Diabetes Risk Category Count")
plt.xlabel("Diabetes Risk")
plt.ylabel("Count")
plt.show()
```
